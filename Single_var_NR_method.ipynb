**Find the roots of x^(3)-x^(2)+2=0, using N-R methods  upto 6th significant digits**

```python
# import necessary modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define lagrange interpolation

def newton(f,df, a, b, e):
  x = (a+b)/2               # initial value
  list_x = [x]
  list_f = [f(x)]
  if f(a)*f(b)>0:
    return [None,list_x,list_f]
  else:
    iter = 0
    h = f(x) / df(x)
    while abs(h)>=e:    
      h = f(x) / df(x)
      x = x - h
      list_x.append(x)
      list_f.append(f(x))
      iter = iter +1
    return [float('%.6g' % x),list_x,list_f]
```

```python
# defining functions
f = lambda x: (x**3) - (x**2) + 2
df = lambda x: (3*(x**2)) - (2*x)

# set the parameters
a = -5
b = 0
e = 0.00001     #error
```

```python
# calculate the root
print ('\n' , 'root is', newton(f,df, a, b, e)[0])
```

```python
# visualization

z = np.arange(a, b, 0.0001)
y = f(z)
plt.plot(z, y, color = 'red', label = 'f')
plt.scatter(newton(f,df, a, b, e)[1],newton(f,df, a, b, e)[2], color = 'blue', label = 'Iterations')
plt.legend()
plt.show()
```
