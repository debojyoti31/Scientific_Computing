**Using Lagrange interpolation formulations  calculate the values of x=301  for  given values of x=300, 304, 305, 307 with f(x)=2.4771, 2.4829, 2.4843 and 2.4871. Correct  upto 6 significant digits through python code.  Also graphically plotted it.**

```python
# import necessary modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define lagrange interpolation
def lagrange_interpolation(x,y,value):
  n = len(x)
  add = 0
  for i in range(n):    
    F = 1   
    for j in range(n):
        if i != j:
            F = F * (value - x[j])/(x[i] - x[j])    
    add = add + F * y[i]
  return round(add, 6)
```

```python
# set the data points
x = [300,304,305,307]
y = [2.4771, 2.4829, 2.4843, 2.4871]

# set the value
value = 301
```

```python
# calculate lagrange interpolation of the value
print("Value at", value, "is", lagrange_interpolation(x,y,value))
```

```python
# visualization
 
x_list = np.arange(np.min(x)-1, np.max(x)+1, 0.001)
z_list = []
for xi in x_list:
  z_list.append(lagrange_interpolation(x,y,xi))
plt.plot(x_list, z_list, color = 'blue', label='lagrange interpolation polynomial, p')
plt.scatter(x, y, color = 'blue', label ='given points')
plt.scatter(value, lagrange_interpolation(x,y,value), color = 'green', label ='value at polynomial p')
plt.legend()
plt.show()
```
